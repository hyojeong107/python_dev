```python
# 박씨네 집 클래스 : 성은 다 '박'
class HousePark:
    lastname = '박' #클래스 변수=> 공유
    # 이름은 상이: 멤버 변수로 처리해야 한다.
    
    def setName(self,name): #풀네임 저장하는 멤버변수 선언하는 메소드 선언 
        self.fullName = HousePark.lastname + name # class에서 직접 접근해서 lastname을 가져옴
        self.fullName = self.lastname + name #오브젝트에 접근해서 lastname을 가져옴
        # self.fullName = lastname + name => 여기서  lastname은 로컬변수
```

```python
aaa = HousePark()
aaa.setName('길동')
print(aaa.fullName)
```

```python
class HousePark:
    lastname = '박'
    
    def setName(self, name):
        self.fullName = HousePark.lastname + name
        
    def travel(self, where): #where이 필요, 이름 필요
        return f'{self.fullName},{where} 여행을 가다'
```

```python
gildong = HousePark()
gildong.setName('길동')
print(gildong.travel('울릉도'))
print(gildong.travel('부산'))
print(gildong.travel('영국'))
```

```python
# 생성자 필요
class HousePark:
    lastName = '박'
    
    #def __init__(self, name):
        #self.fullName = HousePark.lastName+name
        
    def __init__(self, name):
        self.fullName = self.lastName+name # 김씨로 바꿀 수 있다
        
    def travel(self, where):
        return f'{self.fullName}, {where} 여행을 가다'
```

```python
gildong = HousePark('길동')
print(gildong.fullName)
print(gildong.travel('하와이'))
```

```python
# 상속
# 김씨네 집
class HouseKim(HousePark):
    pass
```

```python
# 오버라이딩 : 상속 받은 메소드를 수정해서 사용하겠다 

class HouseKim(HousePark):
    lastName = '김'
    
    def __init__(self, name):
        super().__init__(name)
    
    def travel(self, where):
        return f'{self.fullName}은 {where}(으)로 여행을 아주 신나게 합니다.'
```

```python
gildong = HouseKim('길동')
gildong.travel('부산')
```
