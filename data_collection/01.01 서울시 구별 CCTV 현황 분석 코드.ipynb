```python
# merge라는 함수로
# 병합
# merge()
left = pd.DataFrame({
    'key' : ['k0','k4','k2','k3']
    , 'A' : ['a0','a1','a2','a3']
    , 'B' : ['b0','b1','b2','b3']
})
right = pd.DataFrame({
    'key' : ['k0','k1','k2','k3']
    , 'C' : ['c0','c1','c2','c3']
    , 'D' : ['d0','d1','d2','d3']
})
```

```python
left, right
```

```python
pd.merge( # 특정 컬럼의 값으로 조인 할 때 사용 
    left=left
    , right=right
    , on= 'key' # 조인 할 컬럼명 지정 
    # , how= 'inner' # inner join default
    # , how= 'outer' # outer join default
    # , how= 'left'  # left outer join default
    , how= 'right'   # right inner join default
)
```

```python
# CCTV, 인구현황 merge로 병합 
data_result = pd.merge(
    left = CCTV_Seoul       # left 갈 데이터 프레임 지정 
    ,right = POP_seoul      # right에 갈 데이터 프레임 지정
    , on ='구별'            # 조인조건, 조인 할 컬럼 지정
    , how = 'inner'         # 조인방법 (inner,outer, left outer, right outer)
)
data_result.head()
```

```python
# 직선의 함수 구하기 
# 1. 계수 찾기 f(x) = a*x + b 에서 a,b 찾기 : np.polyfit() 
fp1 = np.polyfit(data_result['인구수'], data_result['소계'],1)
print(fp1)
# 2. 위 계수로 1차 함수 오브젝트 생성: np.polyid(polyfit 결과)
f1 = np.poly1d(fp1)
f1


 
```

```python
f1(100000) , f1(700000)
```

```python
# np.linspace(구간 시작 값,구간 끝 값, 구간 나누어 나온 개수):

for pop in np.linspace(100000,700000, 100): # 10만에서 70만까지 100개로 나누어라 
    
    
    print(pop, ':',f1(pop))
```
