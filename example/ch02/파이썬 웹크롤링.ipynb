```python
# 2.Í∏∞Î≥∏Ï†ÅÏù∏ get ÏöîÏ≤≠
# status_code : ÏùëÎãµÏΩîÎìú(200-Ï†ïÏÉÅÏ≤òÎ¶¨,400-ÏöîÏ≤≠ÏûòÎ™ª,500-ÏÑúÎ≤ÑÏóêÎü¨)
# text : ÏõπÌéòÏù¥ÏßÄÎ•º Î¨∏ÏûêÏó¥Î°ú Î∞òÌôò 
import requests

# ÌÅ¨Î°§ÎßÅ Ìï† ÏõπÏÇ¨Ïù¥Ìä∏ url ÏßÄÏ†ï
url = 'https://www.naver.com' #Î©îÏù∏ ÌéòÏù¥ÏßÄ 

#get ÏöîÏ≤≠ 
response = requests.get(url)

#ÏùëÎãµ ÏΩîÎìú ÌôïÏù∏ 
print('ÏùëÎãµ ÏÉÅÌÉú ÏΩîÎìú: ', response.status_code) #status_code

#ÌéòÏù¥ÏßÄ ÎÇ¥Ïö© Ï∂úÎ†• ÏΩîÎìú(htmlÏΩîÎìú)
print('='*100)
print(response.text)
print('='*100)
print('Îã®Ïñ¥Ïàò: ',len(response.text))
```

```python
# ÏöîÏ≤≠ Ìó§Îçî(User_Agenthu88) Î≥ÄÍ≤ΩÌïòÍ∏∞ 
# ÏÇ¨Ïú†: ÏùºÎ∂Ä ÏõπÏÇ¨Ïù¥Ìä∏Îäî User-AgentÍ∞Ä ÏóÜÎäî ÏöîÏ≤≠ÏùÑ Ï∞®Îã® Ìï† Ïàò ÏûàÎã§ 
# ÏùºÎ∂Ä ÏÇ¨Ïù¥Ìä∏Îäî Í∏∞Î≥∏Ï†ÅÏù∏ Î¥á ÏöîÏ≤≠ÏùÑ Ï∞®Îã®ÌïúÎã§ 
import requests

url = 'https://www.naver.com'

# ÏöîÏ≤≠ Ìó§Îçî ÏÑ§Ï†ï(ÏùºÎ∞òÏ†ÅÏù∏ Î∏åÎùºÏö∞Ï†Ä ÏÇ¨Ïö©ÌïòÎäî Í≤ÉÏ≤òÎüº ÏÑ§Ï†ï)
headers = {
     "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"   
}

#Get ÏöîÏ≤≠ 
response = requests.get(url, headers=headers)

#response ÏÉÅÌÉú ÌôïÏù∏, ÌÖçÏä§Ìä∏ Ï∂îÏ∂ú 
print('ÏùëÎãµÏÉÅÌÉúÏΩîÎìú: ',response.status_code)
print(response.text[:100])
```

```python
# post Î∞©Ïãù ÏöîÏ≤≠ Ï≤òÎ¶¨ÌïòÍ∏∞: insert , update , delete
# requests.post(url,data): dataÎ•º post Î∞©ÏãùÏúºÎ°ú ÏÑúÎ≤ÑÏóê ÎÑòÍ∏¥Îã§.
# url : ÏöîÏ≤≠ Ï£ºÏÜå
# data : ÏÑúÎ≤ÑÎ°ú Î≥¥ÎÇº Îç∞Ïù¥ÌÑ∞ 
# response.json() : ÏÑúÎ≤ÑÏóêÏÑú json Î∞©ÏãùÏúºÎ°ú ÎÑòÏñ¥Ïò® Îç∞Ïù¥ÌÑ∞Î•º ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôòÌï¥ÏÑú Î∞òÌôò 

url = 'https://httpbin.org/post'

# Î≥¥ÎÇº Îç∞Ïù¥ÌÑ∞ 
data = {
    'username':'testuser',
    'password':'testpassword'
}

# post ÏöîÏ≤≠ 
response = requests.post(url,data)
print('ÏùëÎãµÏΩîÎìú: ',response.status_code)
print(response.json()) # Rest API Î∞©ÏãùÏóêÏÑú ÏÑúÎ≤ÑÏóêÏÑú ÎÑòÏñ¥Ïò® Í∞í json
```

```python
# 6. json Îç∞Ïù¥ÌÑ∞ ÏöîÏ≤≠ Î∞è Ï≤òÎ¶¨ 
# Ï¶ùÍ∂åÏÇ¨ api , Rest Api => ÏùëÎãµÏù¥ Îç∞Ïù¥ÌÑ∞ Ïò®Îã§ (json0)

# Í≤åÏãúÌåê 1Î≤àÍ∏Ä ÏöîÏ≤≠ => json ÏùëÎãµ 
url = 'https://jsonplaceholder.typicode.com/posts/1'

response = requests.get(url)

# Ïó¨Í∏∞ÍπåÏßÄ responseÎäî Í≤∞Í≥ºÍ∞Ä json ÌòïÌÉúÏù¥Í∏¥ ÌïòÏßÄÎßå Î¨∏ÏûêÏó¥ 

# ÏúÑ Î¨∏ÏûêÏó¥ÏùÑ Ïö∞Î¶¨Í∞Ä ÎîïÏÖîÎÑàÎ¶¨Î°ú Î≥ÄÌôòÌï¥Ï§òÏïº ÌïúÎã§ 
data = response.json()

print('Ï†úÎ™©: ', data['title'])
print('ÎÇ¥Ïö©: ', data['body'])
```

```python
# ÌÅ¨Î°§ÎßÅÏãú ÏòàÏô∏Ï≤òÎ¶¨ Î∞è Ïò§Î•ò ÎåÄÏùë Î∞©Î≤ï 
# Ïù¥ ÏΩîÎìúÎäî Ïã§Ìñâ ÏïàÎê® 
url = 'https://example.com'

# ÏòàÏô∏Ï≤òÎ¶¨ 
try:
    response = requests.get(url, timeout=5) # ÎåÄÍ∏∞ÏãúÍ∞Ñ(timeout): 5Ï¥àÎèôÏïà ÏùëÎãµÏùÑ Î∞õÍ≤†Îã§
    response.raise_for_status() #raise_for_status() : ÏùëÎãµÏΩîÎìúÍ∞Ä 400Î≤àÎåÄ 500Î≤àÎåÄ Ïùº Îïå ÏòàÏô∏Î∞úÏÉù 
    print(response.text[100])
except requests.Timeout: #ÌÉÄÏûÑÏïÑÏõÉ ÎêòÏóàÏùÑ Îïå Ï≤òÎ¶¨
    print('ÏöîÏ≤≠ÏãúÍ∞ÑÏ¥àÍ≥º') 
except requests.HTTPError as e : #ÎÑ§Ìä∏Ïõç Ïò§Î•ò
    print('HTTPÏò§Î•ò: ',e)
except requests.RequestException as e:
    print('ÏöîÏ≤≠ Ï§ë Ïò§Î•ò Î∞úÏÉù: ',e) # ÏöîÏ≤≠ Ïò§Î•ò
```
