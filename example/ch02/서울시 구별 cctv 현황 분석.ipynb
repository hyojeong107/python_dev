```python
# 컬럼명 수정 : 기관명 => 구별 
# 메소드 : rename()
gu = {
    CCTV_Seoul.columns[0] : '구별'
}
CCTV_Seoul.rename(
    columns=gu
)
```

```python
# rename: 원본이 아니라 복사본이 '구별'로 바뀌었음 
CCTV_Seoul.head()
```

```python
# 컬럼명 변경 : 원본 반영 안되었다.
# 원본에 반영하는 방법 2가지 

# 1. 다시 대입: 기관명 => 구별 
CCTV_Seoul = CCTV_Seoul.rename(columns=gu)
CCTV_Seoul.columns

# 2. inplace=true : 구별 => 기관명 
gu = {
    CCTV_Seoul.columns[0] : '기관명'
}
CCTV_Seoul.rename(columns=gu , inplace=True)
CCTV_Seoul.columns
```
