## request 모듈
    1.모듈소개
        1.requests 모듈: http 요청을 쉽게 보내고 결과를 응답 받는 기능 
            1.요청(requests) : get,post
            2.응답(response)

```python
# 2.기본적인 get 요청
# status_code : 응답코드(200-정상처리,400-요청잘못,500-서버에러)
# text : 웹페이지를 문자열로 반환 
import requests

# 크롤링 할 웹사이트 url 지정
url = 'https://www.naver.com' #메인 페이지 

#get 요청 
response = requests.get(url)

#응답 코드 확인 
print('응답 상태 코드: ', response.status_code) #status_code

#페이지 내용 출력 코드(html코드)
print('='*100)
print(response.text)
print('='*100)
print('단어수: ',len(response.text))
```

```python
# 요청 헤더(User_Agenthu88) 변경하기 
# 사유: 일부 웹사이트는 User-Agent가 없는 요청을 차단 할 수 있다 
# 일부 사이트는 기본적인 봇 요청을 차단한다 
import requests

url = 'https://www.naver.com'

# 요청 헤더 설정(일반적인 브라우저 사용하는 것처럼 설정)
headers = {
     "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"   
}

#Get 요청 
response = requests.get(url, headers=headers)

#response 상태 확인, 텍스트 추출 
print('응답상태코드: ',response.status_code)
print(response.text[:100])
```

```python
# 구글 검색 
# https://www.google.com/search?q=%EB%94%A5%EB%9F%AC%EB%8B%9D&oq=%EB%94%A5%EB%9F%AC%EB%8B%9D&gs_lcrp=EgZjaHJvbWUyDggAEEUYJxg5GIAEGIoFMgcIARAAGIAEMgcIAhAAGIAEMgcIAxAAGIAEMgcIBBAAGIAEMgcIBRAAGIAEMgcIBhAuGIAEMgcIBxAAGIAEMgcICBAAGIAEMgcICRAAGIAE0gEKNzM5ODNqMGoxNagCALACAA&sourceid=chrome&ie=UTF-8

url = 'https://www.google.com/search'

#get 요청에 사용할 파라미터 설정 : 딕셔너리 형태 설정
params = {'q':'딥러닝'}

#요청보내기: get
response = requests.get(url, params=params,headers = { 
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36" 
} 
)

print(response.text)
```

```python
# naver 검색 
query = '머신러닝'
url = f'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={query}'

# get 요청에 사용할 파라미터 설정 : 딕셔너리 형태 설정 
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

#요청 보내기 : get
response = requests.get(url, headers=headers)

print(response.text)
```

```python
# post 방식 요청 처리하기: insert , update , delete
# requests.post(url,data): data를 post 방식으로 서버에 넘긴다.
# url : 요청 주소
# data : 서버로 보낼 데이터 
# response.json() : 서버에서 json 방식으로 넘어온 데이터를 딕셔너리로 변환해서 반환 

url = 'https://httpbin.org/post'

# 보낼 데이터 
data = {
    'username':'testuser',
    'password':'testpassword'
}

# post 요청 
response = requests.post(url,data)
print('응답코드: ',response.status_code)
print(response.json()) # Rest API 방식에서 서버에서 넘어온 값 json
```

```python
# 6. json 데이터 요청 및 처리 
# 증권사 api , Rest Api => 응답이 데이터 온다 (json0)

# 게시판 1번글 요청 => json 응답 
url = 'https://jsonplaceholder.typicode.com/posts/1'

response = requests.get(url)

# 여기까지 response는 결과가 json 형태이긴 하지만 문자열 

# 위 문자열을 우리가 딕셔너리로 변환해줘야 한다 
data = response.json()

print('제목: ', data['title'])
print('내용: ', data['body'])
```

```python
# 크롤링시 예외처리 및 오류 대응 방법 
# 이 코드는 실행 안됨 
url = 'https://example.com'

# 예외처리 
try:
    response = requests.get(url, timeout=5) # 대기시간(timeout): 5초동안 응답을 받겠다
    response.raise_for_status() #raise_for_status() : 응답코드가 400번대 500번대 일 때 예외발생 
    print(response.text[100])
except requests.Timeout: #타임아웃 되었을 때 처리
    print('요청시간초과') 
except requests.HTTPError as e : #네트웍 오류
    print('HTTP오류: ',e)
except requests.RequestException as e:
    print('요청 중 오류 발생: ',e) # 요청 오류
```
